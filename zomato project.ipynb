```python
## importing modules

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## reading csv files using panda

dataframe = pd.read_csv("Zomato-data-.csv")
print(dataframe)
```

```python
## calling dataframe

dataframe     
```

```python
## converting the data type of column - rate
def handleRate(value):
    value=str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
## checking for null value

dataframe.info()
```

```python
## types of resturant

dataframe.head()
```

```python
## checking the majority of resturant falls in which category

sns.countplot(x=dataframe["listed_in(type)"])
plt.xlabel("type of resturant")
```

```python
dataframe.head()
```

```python
## checking which restaurant got the maximum vote

grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes': grouped_data})
plt.plot(result, c="green", marker = "o")
plt.xlabel("Types of Restaurant", c="blue", size=20)
plt.ylabel("votes", c="blue", size=20)
```

```python
## showing rating in histogram to get the range of maximum rating

plt.hist(dataframe['rate'], bins = 5)
plt.title("ratings distributions")
plt.show()
```
